##### Road type and Accident severity

````python
# Define the mapping for numeric road types to their names
road_type_mapping = {
    1: 'Roundabout',
    2: 'One way street',
    3: 'Dual carriageway',
    6: 'Single carriageway',
    7: 'Slip road',
    9: 'Unknown'
}

# Group by road type and accident severity to get the counts
road_severity_counts = df_acci_2020.groupby(['road_type', 'accident_severity']).size().unstack().fillna(0)

# Calculate the percentage of each severity type within each road type
road_severity_percentage = road_severity_counts.div(road_severity_counts.sum(axis=1), axis=0) * 100

# Reset index to make sure the multi-level index is correct
road_severity_percentage = road_severity_percentage.reset_index()

# Map the numeric road types to readable names
road_severity_percentage['road_type'] = road_severity_percentage['road_type'].map(road_type_mapping)

# Now the DataFrame should have readable road type names, and we can print the distribution
print("Accident Distribution by Road Type and Severity:")

# Loop through each road type and print the distribution
for road_type in road_severity_percentage['road_type'].unique():
    print(f"\n{road_type} Accident Distribution:")
    print(road_severity_percentage[road_severity_percentage['road_type'] == road_type].set_index('road_type'))
````

### 1. Are there significant hours of the day, and days of the week, on which accidents occur?

#### Significant hours of the day on which accidents occur

````python
acc_sig_hrs_query = '''
SELECT strftime('%H', time) AS accident_hour,
       COUNT(*) AS accident_count
FROM
    accident WHERE accident_year = 2020
GROUP BY
    accident_hour
ORDER BY
    accident_hour ASC;  -- Sort by accident_hour in ascending order
'''

# Execute the query and fetch results
acc_sig_hrs = cur.execute(acc_sig_hrs_query)
results = acc_sig_hrs.fetchall()
results
````

##### Displaying the counts for hours of the days

````python
# Extract hours and accident counts
hours = [row[0] for row in results]
accident_counts = [row[1] for row in results]

# Plotting the results
plt.figure(figsize=(8, 4))
plt.plot(hours, accident_counts, marker='o', linestyle='-', color='purple')  # Purple line
plt.xlabel('Hour of the Day')
plt.ylabel('Accident Count')
plt.title('Accidents by Hour of the Day')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('Accidents by Hour of the Day.png')
plt.show()
````

#### Significant days of the week on which accidents occur

````python
acc_sig_days_query = '''
SELECT 
    CASE day_of_week
    WHEN 1 THEN 'Sunday'
    WHEN 2 THEN 'Monday'
    WHEN 3 THEN 'Tuesday'
    WHEN 4 THEN 'Wednesday'
    WHEN 5 THEN 'Thursday'
    WHEN 6 THEN 'Friday'
    WHEN 7 THEN 'Saturday'
    END AS accident_day,
    COUNT(*) AS accident_count
FROM 
    accident WHERE accident_year = 2020
GROUP BY 
    day_of_week
ORDER BY
    day_of_week ASC;  -- Sort by day_of_week to keep days in the correct order
'''

# Execute the query and fetch results
acc_sig_days = cur.execute(acc_sig_days_query)
results1 = acc_sig_days.fetchall()
results1
````

##### Displaying accidents counts by days of the week

````python
# Extract days and accident counts
days = [row[0] for row in results1]
accident_counts = [row[1] for row in results1]

# Plotting the results
plt.figure(figsize=(8, 4))
plt.bar(days, accident_counts, color='blue')
plt.xlabel('Day of the Week')
plt.ylabel('Accident Count')
plt.title('Accidents by Day of the Week')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('Accidents by Day of the Week.png')
plt.show()
````

### 4. Using the apriori algorithm, explore the impact of selected variables on accident severity.

````python
# Query the relevant data from the accident table
acc_query = """
    SELECT accident_severity, light_conditions, weather_conditions, road_type, road_surface_conditions
    FROM accident
    WHERE accident_severity IS NOT NULL 
    AND light_conditions IS NOT NULL 
    AND weather_conditions IS NOT NULL
    AND road_type IS NOT NULL
    AND road_surface_conditions IS NOT NULL
     AND accident_year = 2020  -- Filter for the year 20
    LIMIT 10000;
"""
accident_data = pd.read_sql(acc_query, con)

````

````python
# Map the encoded values to meaningful categories
accident_data['accident_severity'] = accident_data['accident_severity'].map({1: 'Severity_Fatal', 2: 'Severity_Serious', 3: 'Severity_Slight'})
accident_data['light_conditions'] = accident_data['light_conditions'].map({1: 'Daylight', 4: 'Darkness_Street_Lights_Present_n_Lit', 5: 'Darkness_Street_Lights_Present_but_UnLit', 6: 'Darkness_No_Street_Lighting', 7: 'Street_lighting_unknown' })
accident_data['weather_conditions'] = accident_data['weather_conditions'].map({1: 'Fine_without_wind', 2: 'Raining_without_wind', 3: 'Snowing_without_wind', 4: 'Fine_with_high_wind', 5: 'Raining_with_high_wind', 6: 'Snowing_with_high_wind', 7: 'Fog_or_Mist_if_hazard', 8: 'Other', 9: 'Unknown'})
accident_data['road_type'] = accident_data['road_type'].map({1: 'Roundabout', 2: 'One_way_street', 3: 'Dual_Carriageway', 6: 'Single_Carriageway', 7: 'Slip_Road', 9: 'Unknown'})
accident_data['road_surface_conditions'] = accident_data['road_surface_conditions'].map({1: 'Road_Dry', 2: 'Road_Wet', 3: 'Snow', 4: 'Frost/Ice', 5: 'Flood'})
````

````python
from mlxtend.frequent_patterns import apriori, association_rules

# One-hot encoding of categorical data
onehot_df = pd.get_dummies(accident_data)

# Run Apriori to find frequent itemsets
frequent_itemsets = apriori(onehot_df, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, num_itemsets=len(frequent_itemsets), metric="confidence", min_threshold=0.6)
````

````python
query_top_3_lsoas = """
SELECT 
    l.lsoa01nm as lsoa_name,  
    l.lsoa01cd as lsoa_code,  
    COUNT(*) as accident_count
FROM accident a
JOIN lsoa l ON a.lsoa_of_accident_location = l.lsoa01cd
WHERE l.lsoa01nm LIKE 'Kingston upon Hull%'
    AND substr(a.date, -4) = '2020'  -- Year is 2020
    AND substr(a.date, 4, 2) IN ('01', '02', '03')  -- Months 01, 02, 03
GROUP BY l.lsoa01cd, l.lsoa01nm
ORDER BY accident_count DESC
LIMIT 3;
"""

# Execute query and create a clear display of results
cur.execute(query_top_3_lsoas)
top_3_lsoas = cur.fetchall()

print("Top 3 Kingston upon Hull LSOAs by Accident Count (January-March 2020)")
print("="*80)
print(f"{'LSOA Name':<30} {'LSOA Code':<15} {'Accident Count':>15}")
print("-"*80)
top_3_lsoas
````

````python
def get_historical_data(lsoa_code):
    query_historical = """
    SELECT 
        date,
        COUNT(*) as daily_accidents
    FROM accident
    WHERE lsoa_of_accident_location = ?
    AND substr(date, -4) = '2020'  -- Year is 2020
    AND substr(date, 4, 2) IN ('01', '02', '03', '04', '05', '06')  -- First 6 months
    GROUP BY date
    ORDER BY 
        substr(date, -4),  -- Year
        substr(date, 4, 2),  -- Month
        substr(date, 1, 2)   -- Day
    """
    return pd.read_sql_query(query_historical, con, params=[lsoa_code])
````

#### Test for Stationarity of the series and modelling

````python
from statsmodels.tsa.stattools import adfuller


def test_stationarity(time_series):
    # Apply Augmented Dickey-Fuller test
    result = adfuller(time_series)
    adf_statistic, p_value = result[0], result[1]
    print(f"ADF Statistic: {adf_statistic}")
    print(f"P-Value: {p_value}")
    
    # If p-value is less than 0.05, the series is stationary
    if p_value < 0.05:
        print("The series is stationary.")
    else:
        print("The series is non-stationary.")
````

#### 8. Construct a social network using the provided data and visualise the network, then provide the basic network characteristics, including numbers of nodes and edges, network density, average degree.

````python
import networkx as nx

# Query a smaller dataset
casualty_data = pd.read_sql("SELECT accident_index, casualty_reference FROM casualty WHERE accident_year = 2020 LIMIT 500;", con)
vehicle_data = pd.read_sql("SELECT accident_index, vehicle_reference FROM vehicle WHERE accident_year = 2020 LIMIT 500;", con)
accident_data = pd.read_sql("SELECT accident_index, lsoa_of_accident_location FROM accident WHERE accident_year = 2020 LIMIT 500;", con)

# Combine datasets efficiently
casualty_vehicle_data = pd.merge(casualty_data, vehicle_data, on="accident_index", how="inner")
casualty_vehicle_location_data = pd.merge(casualty_vehicle_data, accident_data, on="accident_index", how="inner")

# Create the graph
G = nx.Graph()

# Add nodes
G.add_nodes_from(["C_" + str(casualty) for casualty in casualty_vehicle_location_data["casualty_reference"]], type="casualty")
G.add_nodes_from(["V_" + str(vehicle) for vehicle in casualty_vehicle_location_data["vehicle_reference"]], type="vehicle")
G.add_nodes_from(casualty_vehicle_location_data["lsoa_of_accident_location"].unique(), type="location")

# Add edges
G.add_edges_from([
    ("C_" + str(row.casualty_reference), "V_" + str(row.vehicle_reference))
    for _, row in casualty_vehicle_location_data.iterrows()
])
G.add_edges_from([
    ("V_" + str(row.vehicle_reference), row.lsoa_of_accident_location)
    for _, row in casualty_vehicle_location_data.iterrows()
])
G.add_edges_from([
    ("C_" + str(row.casualty_reference), row.lsoa_of_accident_location)
    for _, row in casualty_vehicle_location_data.iterrows()
])

# Visualize the network
plt.figure(figsize=(9, 7))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=False, node_size=20, alpha=0.7)
plt.title("Social Network of Accidents")
plt.savefig("Social Network of Accidents.png")
plt.show()

# Compute basic characteristics
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
density = nx.density(G)
avg_degree = sum(dict(G.degree()).values()) / num_nodes

print(f"Number of Nodes: {num_nodes}")
print(f"Number of Edges: {num_edges}")
print(f"Network Density: {density:.4f}")
print(f"Average Degree: {avg_degree:.2f}")
````
